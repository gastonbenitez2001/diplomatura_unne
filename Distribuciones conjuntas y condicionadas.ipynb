```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Generar datos aleatorios siguiendo una distribución normal
np.random.seed(123)
x = np.random.normal(loc=10, scale=2, size=200)
x = np.sort(x)  # Ordenar los datos

# Calcular la función de densidad de probabilidad (PDF) para los datos
x_pdf = stats.norm.pdf(x, loc=x.mean(), scale=x.std())

# Calcular el área bajo la curva usando la regla del trapecio
area = np.trapz(x_pdf, x)
print(f"El área bajo la curva es: {area:.2f}")

# Graficar el histograma de los datos observados
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.hist(x, bins=20, edgecolor="white", density=True, alpha=0.6, color='g')
plt.plot(x, x_pdf, label='PDF', color='r')
plt.xlabel('Valores de x')
plt.ylabel('Densidad')
plt.title('Histograma de x y PDF')
plt.legend()

# Graficar el histograma de los valores de la PDF
plt.subplot(1, 2, 2)
plt.hist(x_pdf, bins=20, edgecolor="white", density=True, alpha=0.6, color='b')
plt.xlabel('Valores de x_pdf')
plt.ylabel('Densidad')
plt.title('Histograma de x_pdf')

plt.tight_layout()
plt.show()
```
